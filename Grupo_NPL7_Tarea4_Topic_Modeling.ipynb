### Preparado de funciones para limpieza de corpus y tokenización

```python
def tokenize(text):
    tokens=[word for word in nltk.word_tokenize(text) if (len(word) >1 and len(word.strip("-,.?")) >0)]
    #stems=[stemmer.stem(item) for item in tokens]
    return tokens
```

### Selección de hiperparámetros en TFIDFVectorizer:
### Se utilizaron los siguientes parámetros:  stop words "english", min_df=50, max=0.75, max_features= 10,000, token pattern, ngram_range (1,2). 

```python
vectorizer_tf= TfidfVectorizer(tokenizer=tokenize,
                               stop_words='english',
                               max_df=0.75,
                               min_df=100,
                               ngram_range=(1,2),
                               max_features=10000,
                               token_pattern ='(?<!\S)[A-Za-z]+(?!\S)|(?<!\S)[A-Za-z]+(?=:(?!\S))',
                               use_idf=False,
                               norm=None)
```

```python
tf_vectors=vectorizer_tf.fit_transform(X_train.ABSTRACT) 
```

```python
tf_vectors.A
```

```python
vectorizer_tf.get_feature_names()
```

### Aplicación de Modelo LDA Latent Dirichlet Allocation

### Selección de hiperparámetros. Se consideró como hiperparámetro clave a usar en el modelo la definición del número de tópicos de salida que dado el csv original se tienen 6 categorías de clasificación de los documentos. Por lo tanto definimos 6 tópicos para la salida de nuestro modelo (n_components=6).


```python
lda=decomposition.LatentDirichletAllocation(n_components=6,max_iter=3,learning_method='online',learning_offset=50,n_jobs=-1,random_state=111)

W1= lda.fit_transform(tf_vectors)
H1= lda.components_
```

```python
W1
```

```python
H1
```

### Selección de hiperparametros de Topic Modeling. Se consideró en especial un num_words de 5 porque observamos que eran suficientes tokens para entender el significado del tópico.


```python
num_words=5
vocab= np.array(vectorizer_tf.get_feature_names_out())

top_words = lambda t: [vocab[i] for i in np.argsort(t)[:-num_words-1:-1]]
topic_words = ([top_words(t) for t in H1])
topics = [' '.join(t) for t in topic_words]
```

### Clasificación de Tópico Dominante por documento

```python
colnames = ["Topic" + str(i) for i in range(lda.n_components)]
docnames= ["Doc" + str(i) for i in range(len(X_train.ABSTRACT))]
df_doc_topic_train = pd.DataFrame(np.round(W1, 2),columns=colnames, index=docnames)
significant_topic = np.argmax(df_doc_topic_train.values,axis=1)
df_doc_topic_train['dominant_topic']=significant_topic
```

```python
df_doc_topic_train
```

### Aplicación del model LDA en los datos de prueba

```python
WHold = lda.transform(vectorizer_tf.transform(X_hold.ABSTRACT))
```

```python
colnames = ["Topic" + str(i) for i in range(lda.n_components)]
docnames= ["Doc" + str(i) for i in range(len(X_hold.ABSTRACT))]
df_doc_topic_test = pd.DataFrame(np.round(WHold, 2),columns=colnames, index=docnames)
significant_topic = np.argmax(df_doc_topic_test.values,axis=1)
df_doc_topic_test['dominant_topic']=significant_topic
```
